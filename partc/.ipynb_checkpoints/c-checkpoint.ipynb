```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
#load data
raw_data = pd.read_csv("data.csv")

#split into X and y
X = sm.add_constant(raw_data.iloc[:, :-1])
y = raw_data.iloc[:,-1]

#split into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.126, random_state=11)
```

```python
#train the regression model
model = sm.OLS(y_train, X_train)
results = model.fit()

#predict on training data and get R^2
y_train_pred = results.predict(X_train)
print(f"Training R-Squared: {r2_score(y_train, y_train_pred)}")

#predict on testing data and get R^2
y_test_pred = results.predict(X_test)
print(f"Testing R-Squared: {r2_score(y_test, y_test_pred)}")
```

```python
#visualization of training data
plt.scatter(y_train_pred, y_train)
plt.xlabel("Predicted Strength")
plt.ylabel("Actual Strength")
plt.title("Training Data Visualization")
plt.show()

#visualization of testing data
plt.scatter(y_test_pred, y_test)
plt.xlabel("Predicted Strength")
plt.ylabel("Actual Strength")
plt.title("Testing Data Visualization")
plt.show()
```

```python
print(results.summary())
```

```python
#get the new X and y
X = raw_data.iloc[:, :-1]
y = raw_data.iloc[:,-1]

#preprocess
log_trans= ["Age (day)"]
for col in X.columns:
    if col in log_trans:
        X[col] = np.log(X[col])
    else:
        X[col] = (X[col] - X[col].mean()) / X[col].std()

#add the constant
X = sm.add_constant(X)

#split into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.126, random_state=11)

#train the regression model
model = sm.OLS(y_train, X_train)
results = model.fit()

#predict on training data and get R^2
y_train_pred = results.predict(X_train)
print(f"Training R-Squared: {r2_score(y_train, y_train_pred)}")

#predict on testing data and get R^2
y_test_pred = results.predict(X_test)
print(f"Testing R-Squared: {r2_score(y_test, y_test_pred)}")

print(results.summary())
```
